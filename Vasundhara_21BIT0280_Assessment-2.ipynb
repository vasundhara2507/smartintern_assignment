```python
import seaborn as sns #importing important libraries 
```

```python
print(sns.get_dataset_names()) #printing all the dataset available
```

```python
a=sns.load_dataset('car_crashes') #loading 'car_crashes' dataset
```

```python
a #printing 'car_crashes' dataset
```

```python
Inference:
    total -> Number of drivers involved in fatal collisions per billion miles
    speeding -> Percentage Of Drivers Involved In Fatal Collisions Who Were Speeding
    alcohol -> Percentage Of Drivers Involved In Fatal Collisions Who Were Alcohol-Impaired
    not_distracted -> Percentage Of Drivers Involved In Fatal Collisions Who Were Not Distracted
    no_previous -> Percentage Of Drivers Involved In Fatal Collisions Who Had Not Been Involved In Any Previous Accidents
    ins_premium -> Car Insurance Premiums 
    ins_losses -> Losses incurred by insurance companies for collisions per insured driver 
    abbrev -> USA states
```

```python
sns.displot(a['total']) #printing a basic graph 
```

```python
Inference:
    -> The above graph is count with respect to total no. of drivers
    -> The graph above displays the count of states with a specific range of total drivers involved in accidents.
```

```python
sns.scatterplot(x="total",y="speeding",data=a) #printing the data of speeding with respect to total
```

```python
Inference:
    -> The graph shows that they neutrally correlated as we cannot determine it is positive or negative and since 
    correaltion is greater than 0.5 i.e 0.611 so highly correlated.
    -> There appears to be a positive relationship between the total number of drivers involved in fatal collisions 
    per billion miles (Total) and the percentage of drivers involved in those collisions who were speeding (Speeding), 
    suggesting that as the total number of drivers in fatal collisions increases, so does the tendency for speeding to 
    be a contributing factor but however, the correlation is not very strong, as there are some outliers and variations in 
    the data. For example, Montana has a very high number of speeding car crashes (31.29 per billion miles) compared to 
    its total number of car crashes (21.4 per billion miles), while Massachusetts has a very low number of speeding car 
    crashes (4.08 per billion miles) compared to its total number of car crashes (8.2 per billion miles).
```

```python
sns.scatterplot(x="total",y="alcohol",data=a) #printing the data of alcohol with respect to total
```

```python
Inference:
    -> The graph shows that they neutrally correlated as we cannot determine it is positive or negative and since 
    correaltion is greater than 0.5 i.e. 0.85 alcohol so highly correlated.
    -> As the total number of drivers involved in fatal collisions per billion miles (Total) increases, there is a tendency 
    for a higher percentage of drivers involved in those collisions to be alcohol-impaired (Alcohol), indicating a potential 
    positive correlation between these variables but the correlation is not very strong, as there are some outliers and 
    variations in the data. For example, Utah has a very low number of alcohol-related car crashes (1.12 per billion miles) 
    compared to its total number of car crashes (9.62 per billion miles), while North Dakota has a very high number of 
    alcohol-related car crashes (11.01 per billion miles) compared to its total number of car crashes (23.9 per billion miles).
    
    Comparing Both above graph:
        -> Comparing the inferences, it appears that there is a stronger positive correlation between the total number of 
        drivers involved in fatal collisions (Total) and the percentage of drivers who were speeding (Speeding) than there 
        is between Total and the percentage of drivers who were alcohol-impaired (Alcohol). This suggests that speeding 
        might have a more significant influence on fatal collisions in this dataset.
```

```python
sns.set_theme(style="dark")
sns.relplot(x="total",y="alcohol",hue="speeding",size="no_previous",data=a) #setting the theme
```

```python
Inference:
    -> The lightest-colored data points seem to be fading in the white background. So we add the theme so it can be visible.
    -> Themes can be among ('white','dark','whitegrid','darkgrid','ticks') 
```

```python
sns.set_theme(style="darkgrid")
sns.relplot(x="total",y="alcohol",hue="speeding",size="no_previous",marker="v",data=a) #changing marker by giving input for it
```

```python
Inference:
    -> The default marker is a blue circle, as you have already seen in the above output. But the maskers are customizable. 
    This is done by using the marker parameter to set marker types and style.
    -> Marker parameter can be among ('v', '^', '>', '<', 'o', '8', 's', 'p', '*', '+', 'h', 'H', 'D', 'd', 'P', 'X') 
```

```python
cor = a.corr(numeric_only=True)
cor                            #printing correaltion values >0.5 high correlated,<0.5 less correlated and <0 negative correlated
```

```python
sns.heatmap(cor,annot=True) #printing heatmap of correlation
```

```python
Inference:
    -> The heatmap function generates a colour-coded matrix that illustrates how strongly two pairs of variables in a 
    dataset correlate with one another. 
    -> Annot is responsible for writing the correaltion value on the map.
```

```python
sns.heatmap(cor,annot=True,cmap="YlGnBu") #printing heatmap of correlation with specified color
```
